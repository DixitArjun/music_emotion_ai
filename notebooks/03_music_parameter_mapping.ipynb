```python
import pandas as pd
vad_df = pd.read_csv('../data/NRC_VAD/NRC-VAD-Lexicon.txt', sep='\t', header=0)
print(vad_df.head())
print(f"Total words: {len(vad_df)}")
```

```python
def get_vad(sentence):
    words = sentence.split()
    vad_values = []
    
    for word in words:
        row = vad_df[vad_df['term'] == word]
        if not row.empty:
            valence = row['valence'].values[0]
            arousal = row['arousal'].values[0]
            dominance = row['dominance'].values[0]
            vad_values.append((valence, arousal, dominance))
    
    if vad_values:
        avg_valence = sum(v[0] for v in vad_values) / len(vad_values)
        avg_arousal = sum(v[1] for v in vad_values) / len(vad_values)
        avg_dominance = sum(v[2] for v in vad_values) / len(vad_values)
        return [avg_valence, avg_arousal, avg_dominance]
    else:
        return None
#[Valence, Arousal, Dominance]

#Test
result = get_vad("I feel fresh and energetic")
print(result)
```

```python
#Max=160
#Min=50
#Function=Min+((Max-Min)*Arousal)
def get_tempo(arousal):
    return 50+((110)*arousal)
```

```python
get_tempo(result[1])
```

```python
def Arousal_map(lis):
    return get_tempo(lis[1])
```

```python
Arousal_map(result)
```

```python
#Minor or Major Key
def Valence_map(lis):
    return 'minor' if lis[0] <= 0.5 else 'major'
```

```python
Valence_map(result)
```

```python
def get_velocity(arousal,dominance):
    l=(arousal+dominance)/2
    return 30+(80*l)
```

```python
get_velocity(result[1],result[2])
```

```python
def get_nps(arousal):
    return 0.5+(3.5*arousal)
```

```python
def get_pitch(valence,dominance):
    l=(valence+dominance)/2
    return (36+((l)*26),(36+((l)*26))+ 16)
```

```python
get_pitch(0.1,0.1)
```

```python
def get_all(VadList,duration=20):
    tempo = get_tempo(VadList[1])
    key = Valence_map(VadList)
    velocity = get_velocity(VadList[1],VadList[2])
    nps = get_nps(VadList[1])
    pitch = get_pitch(VadList[0],VadList[2])
    
    return {
        'tempo': tempo,
        'key': key,
        'velocity': velocity,
        'note_density': nps,
        'pitch_range': pitch,
        'duration': duration
    }
```

```python
#Happy
get_all([0.94, 0.78, 0.72])
```

```python
#Angry
get_all([0.15, 0.89, 0.75])
```

```python
#Sad
get_all([0.12, 0.21, 0.18])
```

```python
#Peaceful
get_all([0.78, 0.15, 0.55])
```
